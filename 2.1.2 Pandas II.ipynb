```python
# Series -> We had only index name
# Data Frame has both index name and column name
# DataFrame has both row and column
# DataFrame as 2D Array, tabular data
```

```python
# How to create a dataframe
import pandas as pd

data = {
    'Name':['Alice','Bob', 'Charlie'],
    'Salary':[10000,20000,30000],
     'Age':[45,30,35]}

df = pd.DataFrame(data)
df
```

```python
data = {
    'City':['Newyork','London','Paris'],
    'cost':[1000,1200,1500],
    'rate':[5,4,4.5]

}
df = pd.DataFrame(data)
df
```

```python
# Create a dataframe using list

data = [['Newyork','London','Paris'],[1000,1200,1500],[5,4,4.5]]
df = pd.DataFrame(data, columns = ['city','cost','rate'])
df
```
