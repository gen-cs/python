```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)   # Plot the graph
plt.show()      #Show the graph
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)                      # Plot the graph
plt.title("Plot between x and y")  # Add a Title
plt.xlabel("Age")                  # Add a lablel on x axis
plt.ylabel("Salary(in thousands)") # Add a lable on your axis
plt.grid(True)                     # Shows the grid line
plt.show()                         # Show the graph
```

```python
# Add a colour, Line style and Markers

x = [1,2,3,4,5]
y = [1,2,3,4,5]

# Plot the graph
plt.plot(x,y, color = 'green', linestyle = '--', marker = 'o')                      
plt.title("Plot between x and y")  # Add a Title
plt.xlabel("Age")                  # Add a lablel on x axis
plt.ylabel("Salary(in thousands)") # Add a lable on your axis
plt.grid(True)                     # Shows the grid line
plt.show()                         # Show the graph
```

```python
# Add a Legend

x =  [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [3,6,9,12,15]


plt.plot(x,y1, label = 'Line 1', color = 'green', linestyle = '--', marker = '*') # Plot x and y1
plt.plot(x,y2, label = 'Line 2', color = 'red', linestyle = '--', marker = '*')   # Plot x and y2
plt.title("Plot between x and y")  # Add a Title
plt.xlabel("Age")                  # Add a lablel on x axis
plt.ylabel("Salary(in thousands)") # Add a lable on your axis
plt.grid(True)                     # Shows the grid line
plt.legend()                       # Show legend
plt.show() 
```

```python
# Multiple Sub Plot

x =  [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [3,6,9,12,15]

plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title('plot 1')

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('plot 2')

plt.tight_layout()
plt.show()
```

```python
# I have individual plots of x&y1, x&y2, x&y3 but in a same row

# 1 = Row
# 3 = Column

x =  [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [3,6,9,12,15]
y3 = [1,2,3,3,3]
y4 = [1,2,3,4,5]
plt.subplot(1,3,1)
plt.plot(x,y1)
plt.title('plot 1')

plt.subplot(1,3,2)
plt.plot(x,y2)
plt.title('plot 2')

plt.subplot(1,3,3)
plt.plot(x,y2)
plt.title('plot 3')

plt.tight_layout()
plt.show()
```

```python
# I have individual plots of x&y1, x&y2, x&y3 but in a same row

x =  [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [3,6,9,12,15]
y3 = [1,2,3,3,3]
y4 = [1,2,3,5,7]


# 2 = Row
# 2 = Column

plt.figure(figsize = (8,12)) # set the figure size

plt.subplot(2,2,1)
plt.plot(x,y1, label=' L1')
plt.title('plot 1')

plt.subplot(2,2,2)
plt.plot(x,y2, label='L2')
plt.title('plot 2')

plt.subplot(2,2,3)
plt.plot(x,y3, label='L3')
plt.title('plot 3')

plt.subplot(2,2,4)
plt.plot(x,y4, label='L4')
plt.title('plot 4')



plt.show()
```

## Bar Plot

```python
categories = ['A','B','C','D','E']
values = [10,20,30,40,50]

plt.bar(categories,values)
plt.title ('Bar Chart')
plt.show
```

```python
x =  [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [3,6,9,12,15]
y3 = [1,2,5,4,3]
y4 = [1,2,3,5,7]

plt.figure(figsize = (8,12)) # set the figure size



plt.subplot(2,2,1)
plt.bar(x,y1, label=' L1')
plt.title('plot 1')

plt.subplot(2,2,2)
plt.bar(x,y2, label='L2')
plt.title('plot 2')

plt.subplot(2,2,3)
plt.bar(x,y3, label='L3')
plt.title('plot 3')

plt.subplot(2,2,4)
plt.bar(x,y4, label='L4')
plt.title('plot 4')

plt.show
```

```python
categories = ['A','B','C','D','E']
values = [10,20,30,40,50]

plt.bar(categories,values, color ='red')
plt.xlabel('categories')
plt.ylabel('values')
plt.title ('Bar Chart')
plt.grid(linestyle = '--',alpha = 0.5)
plt.show
```

```python
# Horizantal par plot

categories = ['A','B','C','D','E']
values = [10,20,30,40,50]

plt.barh(categories,values, color ='red')
plt.xlabel('categories')
plt.ylabel('values')
plt.title ('Bar Chart')
plt.grid(linestyle = '--',alpha = 0.5)
plt.show
```

# Histogram

```python
import numpy as np

data = np.random.randn(1000)
plt.hist(data, bins = 30, color = 'skyblue',label = 'Plot')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.grid(linestyle = '--',alpha = 0.5)
plt.show()
```

# Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y, color = 'pink', label = 'Virus')
plt.title('Scatter plot')
plt.xlabel('X values')
plt.ylabel('y values')
plt.legend()
plt.grid(linestyle = '--',alpha = 0.1)
plt.show()
```

# Piechart

```python
sizes = [10,40,20,30]
name = ['Apple','Banana','Orange','Cherry']

plt.pie(sizes, labels = name)
plt.title('Pie chart of fruits')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # to "explode" or offset the 2nd slice for emphasis

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle.

plt.show()
```

```python
sizes = [10,40,20,30]
name = ['Apple','Banana','Orange','Cherry']

plt.figure(figsize = (7,7))
plt.pie(sizes, labels = name, autopct='%1.f%%', )
plt.title('Pie chart of fruits')
plt.legend( loc = 'upper right')
plt.show()
```

# Box Plot

# Fill Between

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x) + 0.9

plt.plot(x,y1, color = 'red')
plt.plot(x,y2, color = 'blue')
plt.fill_between(x,y1,y2, color = 'skyblue', alpha = 0.5)
plt.show()
```
