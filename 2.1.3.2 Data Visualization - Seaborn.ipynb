```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Scatter plot
df = sns.load_dataset("tips")
sns.scatterplot(data = df, x = 'total_bill', y = 'tip')
plt.title('Scatter Plot b/w total_bill and tip')
plt.show
```

```python
# Line plot
sns.lineplot(data = df, x = 'size', y = 'tip')
plt.title('Line plot')
plt.show
```

```python
# Bar plot
sns.barplot(data = df, x = 'day', y = 'total_bill')
plt.title('Bar plot')
plt.show
```

```python
# Count Plot
sns.countplot(data=df, x = 'day')
plt.title('Count Plot')
plt.show
```

```python
# Histogram
# kde -> kernel density estimation - smoothed version of histogram
sns.histplot(data=df , x ='total_bill',bins =20, kde = True)
plt.title('Histogram')
plt.show
```

```python
# Box plot
sns.boxplot(data = df, x= 'day',y='total_bill')
plt.title('Box Plot')
plt.show
```

```python
# Violin plot -> as a combination of box plot and kde plot
sns.violinplot(data=df,x='day',y='total_bill')
plt.title('Violin Plot')
plt.show
```

```python
# Strip Plot -> Scatter plot with data points plotted along a categorical axis
sns.stripplot(data = df, x = 'day', y= 'total_bill',jitter = True)
plt.title('Strip plot')
plt.show
```

```python
# Categorical plot
sns.catplot (data = df,x='day', y='total_bill',kind = 'box')
plt.title('Categorical Plot b/w total_bill and day')
plt.show
```

```python
# Correlation + Heatmap
corr = df.corr(numeric_only=True)
print(corr)
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr)
plt.title('Heatmap')
plt.show
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
plt.title('heatmap')
plt.show
plt.show()
```

```python
corr
```

```python
# Cluster map -> is used to cluster similar observations or correlated features
sns.clustermap(corr,annot =True)
plt.title('Cluster map')
plt.show
```

```python
# Pair plot -> it is used to get pairwise relationship btw multiple numeric variables.
df2 = sns.load_dataset("iris")
sns.pairplot(df2,hue = 'species')
plt.title('pair plot')
plt.show
```
