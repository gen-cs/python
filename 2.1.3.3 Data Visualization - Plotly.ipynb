```python
# Import the Plotly package
import plotly.express as px
```

```python
df = px.data.iris()
df
```

```python
# Scatter plot
px.scatter(df, x = 'sepal_width', y = 'sepal_length' , color = 'species', title = 'Scatter plot')
```

```python
# Line chart
df = px.data.gapminder()
df 
# px.line(df,x= 'year	',y = 'gdpPercap',title = 'GDP per Capita')
```

```python
# Line chart
df = px.data.gapminder()
df = df[df['country']== 'India']
df = df[df['country']== 'India']
px.line(df,x= 'year',y = 'gdpPercap',title = 'GDP per Capita for India')
```

```python
df = px.data.gapminder()
df
```

```python
# Bar plot
px.bar(df, x = 'year' , y = 'pop', color = 'continent', title = 'Polulation over time')
```

```python
# Histogram
px.histogram(df,x='lifeExp', nbins = 50, color = 'continent')
```

```python
# Box plot
px.box(df, x ='continent', y= 'lifeExp', color = 'continent')
```

```python
# Pie chart
px.pie(df,names = 'continent', values = 'pop', title = 'continent population share')
```

```python
# Pie chart
px.pie(df,names = 'country', values = 'pop', title = 'continent population share')
```

```python
# Area Chart
px.area(df,x='year',y='pop',color = 'continent', line_group = 'country')
```
